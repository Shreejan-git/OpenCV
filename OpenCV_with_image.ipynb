```python
import cv2
import os
import matplotlib.pyplot as plt
```

```python
img_path = os.path.join('images','tiger.jpg')
img_path
```

```python
cv2.imread??
#using two question marks sign will display the documentation.
```

```python
img = cv2.imread(img_path)
img.shape
#displaying the height, width and channel. Since the img is a colored image.
```

```python
#cv2.imshow() was not working in jupyter notebook so using matplotlib to render the image.
#but if we look at the image, it has displayed differently.
#this is because we read the image using OpenCV, and OpenCV reads an image in the form of BGR (Blue, Green, Red)
#instead of RGB (Red Green Blue)
plt.imshow(img)
plt.show()
```

```python
#to fix the above situation:
recolored_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(recolored_image)
plt.show()
```

to write out the image in different format

```python
new_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imwrite('images/gray_tiger.jpg', new_img)
```
