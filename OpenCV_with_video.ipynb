```python
cap = cv2.VideoCapture('video/video_opencv.mp4')
```

```python
for frame_index in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    #we need to typecast into the return value of cap.get(cv2.CAP_PROP_FRAME_COUNT into int.
    ret, frame = cap.read()
    
    cv2.imshow('video player',frame)
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
